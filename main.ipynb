# Professional visualization to clarify the relationship between WFH and productivity/performance

## Executive summary

The analysis of remote work dynamics in New South Wales (NSW), Australia, through various survey data from 2020 and 2021, yields insightful revelations about the impact of the COVID-19 pandemic on work habits and preferences. The transition to remote work, hours dedicated to work and personal time, and the influence of management practices on infrastructure flexibility and workspace benefits have been explored, leading to several key insights:

- Transition to Remote Work: 

Pre-pandemic, remote work was relatively uncommon across all age groups in NSW, with minimal engagement in 2019. The pandemic-induced shift in 2020 saw a significant rise in remote work, becoming nearly universal across age demographics, highlighting the adaptability of the workforce to unprecedented circumstances.

- Work Hours: 

Contrary to expectations that remote work might lead to increased overtime, the data indicates no substantial change in work hours, suggesting that employees maintained consistent work patterns despite the shift in work environment.

- Personal and Family Time: 

The shift to remote work led to an increase in time spent on personal and family activities, particularly in 2021, reflecting the blurring of work-life boundaries and emphasizing the importance of work-life balance in remote work settings.

- Caring and Domestic Responsibilities: 

The pandemic and associated restrictions contributed to an increase in time dedicated to family duties. However, distinguishing whether this shift is directly attributable to remote work or the broader context of the pandemic remains challenging.

- Workplace and Management: 

Poor management is significantly correlated with workspace satisfaction, with improvements in management potentially enhancing employees' workspace perceptions. Additionally, a correlation between poor management and living situations suggests that flexible work location choices could mitigate workplace barriers.

- Top Management Support (TMS) and Infrastructure Flexibility (INF): 

A moderate correlation between TMS and INF underscores the positive impact of management improvement on infrastructure flexibility, including IT and cybersecurity, essential for effective remote work.

- External Factors and Workplace Benefits: 

Additional costs are notably correlated with workspace quality, indicating that financial barriers can impede the creation of conducive home working environments.

- Remote Work Culture and Employee Satisfaction: 

The culture of remote work significantly influences employees' perceptions of WFH, highlighting the importance of a supportive remote work culture for employee satisfaction and productivity.

- Remote Work and Organizational Recruitment: 

Remote job satisfaction and the perception of remote work positively impact recruitment, suggesting that remote work policies can enhance an organization's attractiveness to potential employees.

- Management Practices and Recruitment: 

There is a notable correlation between effective remote management practices, focusing on results, and the ability to recruit successfully, emphasizing the role of management in leveraging remote work for recruitment advantages.

In conclusion, the analysis suggests that the pandemic has not only forced a shift to remote work but also offered an opportunity to reassess work habits, management practices, and organizational policies to support effective and satisfying remote work environments.

# EDA

Exploratory Data Analysis (EDA) is a crucial step in understanding the dynamics and implications of work from home (WFH) practices, as revealed by the New South Wales (NSW) survey data. This analysis involves a systematic examination of the dataset, which comprises responses from two waves of surveys.

The goal of our EDA is to uncover patterns, trends, and correlations within the data that shed light on how WFH is conducted in NSW in 2018 to early 2021.

```python
q = a.Q(questions,questions_2020,questions_2021,rws_2020,rws_2021)
_ = q.get_qid_type('experience',questions,True)
```

```python
import math
import itertools
import textwrap

def eda(common_qid):

    q_info = q.get_q_info(common_qid)
    df_list = q.one_qid_dfs(common_qid)

    percent_scale = {
    "Rarely or never":"01. 0%",
    "Rarely or never ":"01. 0%",
    "Less than 10% of my time":"02. 1%-10%",
    "10%":"03. 10%",
    "20%":"04. 20%",
    "30%":"05. 30%",
    "40%":"06. 40%",
    "50% - I spent about half of my time remote working":"07. 50%",
    "60%":"08. 60%",
    "70%":"09. 70%",
    "80%":"10. 80%",
    "90%":"11. 90%",
    "100% - I spent all of my time remote working":"12. 100%"
    }

    for target_df in df_list:

        target_df_cols = target_df.columns.tolist()
        target_df_cols.remove(q.id[0])

        num_cols = len(target_df_cols)

        target_col = target_df_cols[0]

        num_subplot_cols = 2
        num_subplot_rows = int(math.ceil(num_cols / num_subplot_cols))

        fig = plt.figure(figsize = (15,5))
        axes = fig.subplots(num_subplot_rows, num_subplot_cols,sharex = 'all',sharey = 'all')
        fig.suptitle('{}'.format(q_info['question']))

        pos = list(itertools.product([r for r in range(num_subplot_rows)],[c for c in range(num_subplot_cols)]))

        for (r,c),target_col in zip(pos,target_df_cols):

            target_df[target_col].replace(percent_scale, inplace = True)

            if num_subplot_rows == 1:
                sns.countplot(data = target_df, y  = target_col,order = np.sort(target_df[target_col].unique()).tolist() ,ax=axes[c])
                axes[c].set_title('{}'.format(textwrap.fill(target_col, width=60)))
                axes[c].set_ylabel(' ')
                axes[c].set_xlabel('{} count'.format(q.id[0]))

            elif num_subplot_rows >= 2:
                sns.countplot(data = target_df, y  = target_col,order = np.sort(target_df[target_col].unique()).tolist() , ax = axes[r,c])
                axes[r,c].set_title('{}'.format(textwrap.fill(target_col, width=60)))
                axes[r,c].set_ylabel('')
                axes[r,c].set_xlabel('{} count'.format(q.id[0]))

            plt.tight_layout()
```

```python
bq = common_comb[0][0]
eq = common_comb[0][1]

def eda2(bq,eq,fsize):
    dfs = q.two_qids_dfs(bq,eq)

    target_df_2020 = dfs['2020']['merged']
    target_df_2021 = dfs['2021']['merged']

    q_info = q.get_q_info(bq)

    bio_q2020 = list(set(target_df_2020.columns) - set(exp_q2020.tolist() + q.id))
    bio_q2021 = list(set(target_df_2021.columns) - set(exp_q2021.tolist() + q.id))

    bio_q = bio_q2020
    exp_q = exp_q2020
    df = target_df_2020

    for i in df.loc[:,exp_q]:

        df.loc[:,exp_q] = df.loc[:,exp_q].replace(percent_scale)

    combs = list(itertools.product(bio_q,exp_q))

    num_subplot_cols = 2
    num_subplot_rows = int(math.ceil(len(combs)/ num_subplot_cols))

    fig = plt.figure(figsize = fsize)
    axes = fig.subplots(num_subplot_rows, num_subplot_cols,sharex = 'all',sharey = 'all')
    fig.suptitle('{}'.format(q_info['question']))

    pos = list(itertools.product([r for r in range(num_subplot_rows)],[c for c in range(num_subplot_cols)]))

    for (r,c),(b,e) in zip(pos,combs):
        
        df_ctab = pd.crosstab(df[b], df[e], values = df[q.id], aggfunc = 'count',normalize = 'index')

        if num_subplot_rows == 1:
            sns.heatmap(df_ctab,annot = True, fmt=".0%", cmap="YlGnBu",ax = axes[c])
            axes[c].set_ylabel('{}'.format(textwrap.fill(b, width=60)))
            axes[c].set_xlabel('{}'.format(textwrap.fill(e, width=60)))

        elif num_subplot_rows >= 2:
            sns.heatmap(df_ctab,annot = True, fmt=".0%", cmap="YlGnBu",ax = axes[r,c])
            axes[r,c].set_ylabel('{}'.format(textwrap.fill(b, width=60)))
            axes[r,c].set_xlabel('{}'.format(textwrap.fill(e, width=60)))

        plt.tight_layout()

eda2(bq,eq,(15,10))
```

## analysis approach

Various analytical approaches are employed to explore the relationships delineated by our conceptual model. Initially, we endeavor to present professional visualizations utilizing classic statistical methods. Given the complexity of our conceptual model, applying Structural Equation Modeling (SEM) to fully understand the correlations among attributes proves challenging. Despite the inability to observe interaction effects, latent factors, or any confounding factors directly, classic statistical tests offer a straightforward means to examine the correlation between two variables. Consequently, we have opted for this approach in our analysis, focusing primarily on data from the second wave (2021).

Our dataset encompasses two categories of variables: quantified and qualified. The qualified variables are further distilled into nominal and ordinal scales. For qualified variables with more than three levels, we also undertake additional tests to discern which levels are correlated. This classification guides the selection of statistical tests and methods for computing correlations. Below, the structure of our analytical method elucidates our process for achieving professional visualizations.

![](img\statistical_approach.png)

Depending on the types of variables involved, the statistical test we apply varies:

- For two quantified variables:

  The simplest scenario involves conducting Welch's T-test to determine correlation. Unlike Before-After comparisons that involve a single respondent, this comparison is independent, necessitating an unpaired test. While the Student T-test or Mann-Whitney U test may sometimes be more suitable, employing multiple tests for a single case is not ideal. Welch's T-test is advantageous as it does not require the assumption of normal distribution or equal variances across populations, making it universally applicable.

- For one quantified and one qualified variable:

  Initially, the Shapiro-Wilk test assesses whether the quantified variable follows a normal distribution. If so, the Bartlett test checks for equality of variances across populations. When a variable has more than three levels, the Shapiro-Wilk test must be applied to multiple arrays, which could lead to misinterpretation due to repeated testing. To mitigate this, we apply the Bonferroni correction to rigorously assess statistical significance by adjusting p-values for the number of tests. If both normal distribution and equal variances are confirmed, we proceed with a one-way ANOVA for more than two groups. Alternatively, the Kruskal-Wallis Test is used when equal variance is not verified. Subsequently, correlation strengths are calculated using either the correlation ratio method or Spearman's Correlation, depending on whether the qualified variable is nominal or ordinal. For variables with more than three levels, we explore correlated levels using Tukey's or Steel-Dwass tests.

- For two qualified variables

  If both are nominal, the Chi-square test is applied. Upon finding a significant association, Cramer's V measures the strength of correlation. If one variable is ordinal, Spearman's Correlation is used. For variables with more than three levels, Chi-square residual analysis identifies the specific cells contributing to the association.

In summary, the choice of statistical tests is dictated by the data type (quantified or qualified) and scale (nominal or ordinal). For quantified variables, we employ tests like Welch's T test or one-way ANOVA, whereas for qualified variables, the selection varies between the Kruskal-Wallis H-test, Chi-square test, or Steel-Dwass test, with a preference for non-parametric tests when data assumptions are not met.

## statistics application

As previously mentioned, this questionnaire comprises approximately 180 questions in total. Analyzing pairs one by one becomes impractical given the number of question pairs is $180C2$, totaling over 16,000 patterns. To efficiently investigate these question pairs, we have developed a statistical application as outlined below. The notebook titled "main" serves as the report file for this analysis, while "main.py" acts as an integration file, implementing all the functions. The process is divided into three main streams: "a_data_process.py", "b_stats_approach.py", and "c_viz.py", executed in sequential order.

File A is responsible for cleaning the survey's raw data, ensuring its consistent usability. This file's components are crucial; for example, it addresses missing value imputation. To maximize the information from the raw data, we opt for univariate imputation over Pair-wise or List-wise deletion. Although multivariate imputation could be more comprehensive, its systematic implementation to universally fill all types of missing values poses challenges. For categorical variables, missing values are imputed with the most frequent value or an 'Empty' text. To minimize data analyst bias, "Empty" text is used to clearly denote a cell as NULL if the missing value ratio exceeds 10%. Conversely, if the ratio is below this threshold, the most frequent value is used for imputation. For numerical variables, we impute missing values with either the mean or median value. Should the outlier ratio surpass 0.3%, we consider their influence significant and opt for median imputation, whereas mean imputation is applied otherwise.

File B constitutes the core of the statistical analysis implementation, and file C facilitates visualization. Associated with file B, several helper functions act as individual statistical methods. For example, when a Bartlett test is required, "equal_variance_test.py" is called from file B.

Moreover, we have predefined configurable parameters stored in YAML files. These files set hyperparameters, such as common question IDs. "Utility.py" includes useful functions for addressing common preprocessing tasks.

```python
import correlation_test as ct
import utility as u
import correlation as cor
import distribution_test as dit
import b_stats_approach as b

q = a.Q(questions,questions_2020,questions_2021,rws_2020,rws_2021)
qids_list = q.get_qid_model("EMS",questions,True)
print_ = False
threshold = .05

result_list = []

for qid in qids_list:
    df_list = q.one_qid_dfs(qid)
    for qdf in df_list:        
        for col in qdf.columns:
            if col != q.id[0]:

                arr1 = qdf[col]
                if isinstance(arr1,pd.DataFrame):
                    arr1 = arr1.iloc[:,0]
        
                arr2 = arr_gap

                if len(arr1) == len(arr2):
        
                    arr_types = [u.istype(arr) for arr in [arr1,arr2]]
                    arr_scales = [u.isscale(arr,likert_scale) for arr in [arr1,arr2]]

                    if print_ == True:
                        print('::::::::::::::::::::::::::::::::')
                        print('arr1 : {}'.format(arr1.name))
                        print('arr2 : {}'.format(arr2.name))
                        print('arr_types : {}'.format(arr_types))
                        print('arr_scales : {}'.format(arr_scales))
                        print('::::::::::::::::::::::::::::::::')

                    if arr_types == ['qualified','qualified']: # ('chi2','crosstab')

                        test_result = ct.crosstab_chi2(arr1,arr2, print_ , threshold, values = qdf.loc[:,q.id], aggfunc = 'nunique') # [p,chi2_result]

                    elif arr_types == ['qualified','quantified']: # ('avg diff test','tab on arr1')
                        
                        arr_list = u.array_split(arr1,arr2)
                        distribution_variance_test_result = dit.distribution_variance_test(print_,threshold,*arr_list)
                        test_result = b.qualified_quantified_test(distribution_variance_test_result,print_,threshold,*arr_list)

                    elif arr_types == ['quantified','qualified']: # ('avg diff test','tab on arr2')

                        arr_list = u.array_split(arr2,arr1)
                        distribution_variance_test_result = dit.distribution_variance_test(print_,threshold,*arr_list)
                        test_result = b.qualified_quantified_test(distribution_variance_test_result,print_,threshold,*arr_list)
                    
                    elif arr_types == ['quantified','quantified']: # ('t test','as is')

                        test_result = ct.t_test(arr1,arr2,'welch_t',print_)

                    corr_result = cor.compute_correlation(arr1,arr2,arr_types,arr_scales,print_)

                    result_list.append(list([arr1.name,arr2.name]) + test_result + list(corr_result))

result_df = pd.DataFrame(result_list,columns = ['question1','question2','test_p-value','test_result','corr','corr_p-value','corr_test'])
main_result = result_df.query('test_result == True & corr > .3')
main_result
```

```python
import correlation_test as ct
import utility as u
import correlation as cor
import distribution_test as dit
import b_stats_approach as b

q = a.Q(questions,questions_2020,questions_2021,rws_2020,rws_2021)
qids_list = q.get_qid_model("EMP",questions,False)
print_ = False
threshold = .05

result_list = []

for qid in qids_list:
    df_list = q.one_qid_dfs(qid)
    for qdf in df_list:        
        for col in qdf.columns:
            if col != q.id[0]:

                arr1 = qdf[col]
                if isinstance(arr1,pd.DataFrame):
                    arr1 = arr1.iloc[:,0]
        
                arr2 = arr_gap

                if len(arr1) == len(arr2):
        
                    arr_types = [u.istype(arr) for arr in [arr1,arr2]]
                    arr_scales = [u.isscale(arr,likert_scale) for arr in [arr1,arr2]]

                    if print_ == True:
                        print('::::::::::::::::::::::::::::::::')
                        print('arr1 : {}'.format(arr1.name))
                        print('arr2 : {}'.format(arr2.name))
                        print('arr_types : {}'.format(arr_types))
                        print('arr_scales : {}'.format(arr_scales))
                        print('::::::::::::::::::::::::::::::::')

                    if arr_types == ['qualified','qualified']: # ('chi2','crosstab')

                        test_result = ct.crosstab_chi2(arr1,arr2, print_ , threshold, values = qdf.loc[:,q.id], aggfunc = 'nunique') # [p,chi2_result]

                    elif arr_types == ['qualified','quantified']: # ('avg diff test','tab on arr1')
                        
                        arr_list = u.array_split(arr1,arr2)
                        distribution_variance_test_result = dit.distribution_variance_test(print_,threshold,*arr_list)
                        test_result = b.qualified_quantified_test(distribution_variance_test_result,print_,threshold,*arr_list)

                    elif arr_types == ['quantified','qualified']: # ('avg diff test','tab on arr2')

                        arr_list = u.array_split(arr2,arr1)
                        distribution_variance_test_result = dit.distribution_variance_test(print_,threshold,*arr_list)
                        test_result = b.qualified_quantified_test(distribution_variance_test_result,print_,threshold,*arr_list)
                    
                    elif arr_types == ['quantified','quantified']: # ('t test','as is')

                        test_result = ct.t_test(arr1,arr2,'welch_t',print_)

                    corr_result = cor.compute_correlation(arr1,arr2,arr_types,arr_scales,print_)

                    result_list.append(list([arr1.name,arr2.name]) + test_result + list(corr_result))

result_df = pd.DataFrame(result_list,columns = ['question1','question2','test_p-value','test_result','corr','corr_p-value','corr_test'])
main_result = result_df.query('test_result == True & corr > .3')
main_result
```
